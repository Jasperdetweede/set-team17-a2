```python
from hill_climbing import generate_line_stripe_neighbour
from keras.utils import load_img, img_to_array
import matplotlib.pyplot as plt
import numpy as np

# ----------------------------
# LOAD IMAGE
# ----------------------------
image_path = "images/cup.jpeg"  # replace with your image
epsilon = 0.3

img = load_img(image_path)
img_array = img_to_array(img)

# ----------------------------
# GENERATE MUTATED IMAGE
# ----------------------------
mutated_img = generate_line_stripe_neighbour(img_array, epsilon, num_lines=10, width=10)

# ----------------------------
# DISPLAY ORIGINAL VS MUTATED
# ----------------------------
plt.figure(figsize=(10,5))

plt.subplot(1,2,1)
plt.title("Original")
plt.imshow(img_array.astype(np.uint8))
plt.axis("off")

plt.subplot(1,2,2)
plt.title("Mutated with line stripes")
plt.imshow(mutated_img.astype(np.uint8))
plt.axis("off")

plt.show()

# ----------------------------
# CHECK MAX DELTA
# ----------------------------
perturbation = np.abs(mutated_img - img_array)
print("Max delta:", perturbation.max())
print("Within epsilon bound:", np.all(perturbation <= 255 * epsilon))
```
